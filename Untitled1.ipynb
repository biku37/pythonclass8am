```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup=requests.get(url)
soup
```

```python
codes=soup.text
codes=BeautifulSoup(codes,'lxml')
```

```python
tag=codes.table.find_all('tr')
for i in tag:
    print(i.text)
```
